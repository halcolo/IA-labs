```python
import numpy as np
import scipy as sc

import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()

print (boston.DESCR)
```

# Formula to minimize Mean Squared Error (MSE)
$\beta = (X^{T}X)^{-1}X^{T}Y$


```python
X =  boston.data[:, 5]
Y = boston.target

plt.scatter(X, Y, alpha=0.3)


# Add ones column in the X matrix
X = np.array([np.ones(506), X]).T

B = np.linalg.inv(X.T @ X) @ X.T @ Y 

print (B)

plt.plot([4, 9], [B[0] + B[1]* 4, B[0] + B[1] * 9], c='red')
plt.show()
```
